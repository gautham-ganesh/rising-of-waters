# Flood Prediction Notebook

This notebook demonstrates dataset inspection, cleaning, feature scaling, and model training guidance for the flood prediction app.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
```

```python
df = pd.read_csv('data/flood_prediction.csv')
df.head()
```

```python
print(df.info())
print('Missing values:')
print(df.isna().sum())
```

```python
feature_columns = ['rainfall_mm', 'river_level_m', 'soil_moisture_pct', 'catchment_area_km2', 'temperature_celsius']
target_column = 'flood_risk'
df[feature_columns + [target_column]] = df[feature_columns + [target_column]].apply(pd.to_numeric, errors='coerce')
clean_df = df.dropna(subset=feature_columns + [target_column])
print('Clean shape:', clean_df.shape)
clean_df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(clean_df['rainfall_mm'], kde=True)
plt.title('Rainfall Distribution')
plt.show()
```
